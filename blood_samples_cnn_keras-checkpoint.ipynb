```python
img_width, img_height = 60,60
batch_size = 32

print(os.listdir("C:\\Users\\tera baap\\Desktop\\TRAIN"))
```

```python
a_images = []
for key in os.listdir(os.path.join(path_to_data, "EOSINOPHIL")):
    if not "TRAIN" in key:
        a_images.append(os.path.join(path_to_data, "EOSINOPHIL", key))
for key in os.listdir(os.path.join(path_to_data,  "LYMPHOCYTE")):
    if not "TRAIN" in key:
        a_images.append(os.path.join(path_to_data, "LYMPHOCYTE", key))
for key in os.listdir(os.path.join(path_to_data,  "MONOCYTE")):
    if not "TRAIN" in key:
        a_images.append(os.path.join(path_to_data,  "MONOCYTE", key))
for key in os.listdir(os.path.join(path_to_data,  "NEUTROPHIL")):
    if not "TRAIN" in key:
        a_images.append(os.path.join(path_to_data, "NEUTROPHIL", key))
        
shuffle(a_images)
print(len(a_images))
```

```python
test_images = []
for key in os.listdir(os.path.join(path_to_data1, "EOSINOPHIL")):
    if not "TEST" in key:
        test_images.append(os.path.join(path_to_data1, "EOSINOPHIL", key))
for key in os.listdir(os.path.join(path_to_data1,  "LYMPHOCYTE")):
    if not "TEST" in key:
        test_images.append(os.path.join(path_to_data1, "LYMPHOCYTE", key))
for key in os.listdir(os.path.join(path_to_data1,  "MONOCYTE")):
    if not "TEST" in key:
        test_images.append(os.path.join(path_to_data1,  "MONOCYTE", key))
for key in os.listdir(os.path.join(path_to_data1,  "NEUTROPHIL")):
    if not "TEST" in key:
        test_images.append(os.path.join(path_to_data1, "NEUTROPHIL", key))
        
shuffle(test_images)
print(len(test_images))
```

```python
path_to_data2 = "C:\\Users\\tera baap\\Desktop\\TEST_SIMPLE"
```

```python
test_2_images = []
for key in os.listdir(os.path.join(path_to_data2, "EOSINOPHIL")):
    if not "TEST_SIMPLE" in key:
        test_2_images.append(os.path.join(path_to_data2, "EOSINOPHIL", key))
for key in os.listdir(os.path.join(path_to_data2,  "LYMPHOCYTE")):
    if not "TEST_SIMPLE" in key:
        test_2_images.append(os.path.join(path_to_data2, "LYMPHOCYTE", key))
for key in os.listdir(os.path.join(path_to_data2,  "MONOCYTE")):
    if not "TEST_SIMPLE" in key:
        test_2_images.append(os.path.join(path_to_data2,  "MONOCYTE", key))
for key in os.listdir(os.path.join(path_to_data2,  "NEUTROPHIL")):
    if not "TEST_SIMPLE" in key:
        test_2_images.append(os.path.join(path_to_data2, "NEUTROPHIL", key))
        
shuffle(test_2_images)
print(len(test_2_images))
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape= (img_width, img_height, 3), name="conv1"))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(32, (3, 3), name="conv2"))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3), name="conv3"))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3, 3), name="conv4"))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128))
model.add(Dropout(0.5))
model.add(Activation('relu'))

model.add(Dense(64))
model.add(Dropout(0.5))
model.add(Activation('relu'))

model.add(Dense(4))
model.add(Activation('sigmoid'))

model.summary()
```

```python
from keras.utils import to_categorical
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
train_samples = len(a_images)
test_samples = len(test_images)
test_2_samples = len(test_2_images)

print("Train samples: " + str(train_samples))
print("test samples: " + str(test_samples))
print("test_2_samples: " + str(test_2_samples))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
     shear_range=0.0001,
    zoom_range=2,
    horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255,
   shear_range=0.0001,
    zoom_range=2,
    horizontal_flip=True)

train_generator = train_datagen.flow_from_directory(
    'C:\\Users\\tera baap\\Desktop\\TRAIN',
    target_size=(60,60),
    batch_size=32,
    class_mode='categorical')

test_generator = test_datagen.flow_from_directory(
    'C:\\Users\\tera baap\\Desktop\\TEST',
    target_size=(60,60),
    batch_size=32,
    class_mode='categorical')

test_2_generator = test_datagen.flow_from_directory(
    'C:\\Users\\tera baap\\Desktop\\TEST_SIMPLE',
    target_size=(60,60),
    batch_size=32,
    class_mode='categorical')
```

```python
model.fit_generator(
    train_generator,
    epochs= 100,
    steps_per_epoch= train_samples/batch_size,
    validation_data= test_generator,
    validation_steps = test_samples/batch_size 
)
```
